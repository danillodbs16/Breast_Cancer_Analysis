# Data Set 1:

## Breast Cancer Wisconsin (Categorical data)

- **Attributes**:
  
  1. Sample code number id number
  2. Clump Thickness 1 - 10
  3. Uniformity of Cell Size 1 - 10
  4. Uniformity of Cell Shape 1 - 10
  5. Marginal Adhesion 1 - 10
  6. Single Epithelial Cell Size 1 - 10
  7. Bare Nuclei 1 - 10
  8. Bland Chromatin 1 - 10
  9. Normal Nucleoli 1 - 10
  10. Mitoses 1 - 10
  11. Class (2 for benign, 4 for malignant)

**Original websites:**
1. https://www.kaggle.com/datasets/mariolisboa/breast-cancer-wisconsin-original-data-set
2. https://archive.ics.uci.edu/dataset/15/breast+cancer+wisconsin+original

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mariolisboa/breast-cancer-wisconsin-original-data-set")

print("Path to dataset files:", path)
```

```python
df1=pd.read_csv(glob.glob(path+"/*")[0])
```

```python
df1.columns
```

### Fixing lack of numerical information from Bare Nuclei

```python
L=[]
for i in df1["Bare Nuclei"]:
    try:
        L.append(int(i))
    except:
        L.append(np.nan)
df1["Bare Nuclei"]=L
```

```python
df1=df1.dropna()
```

### Uniformizing tumour class (2 = Benign, 4 = Malignant)

```python
def fix_class(val):
    if val==2:
        return "Benign"
    else:
        return "Malignant"
    
```

```python
df1["Class"]=df1["Class"].apply(fix_class)
```

```python
df1
```

### Sample code number is not unique!

```python
df1["Sample code number"].nunique()
```

```python
len(df1)
```

### Replacing sample code number by the dataframe numeric indexing:

```python
df1["Sample number"]=[i+1 for i in range(len(df1))]
```

```python
df1=df1.drop(columns="Sample code number")
```

```python
df1=df1[["Sample number"]+list(df1)[:-1]]
```

```python
df1
```

```python
np.random.uniform(0,0.001)
```

```python
for i in df1.columns:
    try:
        df1[i]=[j+np.random.uniform(0,0.0001) for j in df1[i]]
    except:
        pass
```

```python
df1
```

# Dataset 2:

## Breast Cancer - Wisconsin (numerical data)

**Website:** https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data

**Short Report:**

1. Total number of columns: 10;
2. Total number of entries: 699;

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uciml/breast-cancer-wisconsin-data")

print("Path to dataset files:", path)
```

```python
df2=pd.read_csv(glob.glob(path+"/*")[0])
```

```python
df2.info()
```

### Removing null column "Unnamed: 32"

```python
df2=df2.drop(columns="Unnamed: 32")
```

### Changing diagnosis from M and B to Maligmant and Benign, respectively.

```python
def fix_diag(val):
    if val=="M":
        return "Malignant"
    else:
        return "Benign"
```

```python
df2["diagnosis"]=df2["diagnosis"].apply(fix_diag)
```

# Creating Database in the "data" folder

```python
import os
import sqlite3

# --- Step 1: Set the correct relative path to the 'data' folder ---
data_folder = os.path.abspath(os.path.join(os.getcwd(), "../data"))
print(f"Using data folder: {data_folder}")

# --- Step 2: Ensure the folder exists ---
if not os.path.exists(data_folder):
    try:
        os.makedirs(data_folder, exist_ok=True)
        print(f"Created folder: {data_folder}")
    except PermissionError:
        raise PermissionError(f"Cannot create folder: {data_folder}. Check permissions.")

# --- Step 3: Check write access ---
if not os.access(data_folder, os.W_OK):
    raise PermissionError(f"No write permission for folder: {data_folder}")

# --- Step 4: Build DB path ---
db_path = os.path.join(data_folder, "BC_database_final_version_mod.db")

# --- Step 5: Connect to SQLite and save DataFrames ---
try:
    conn = sqlite3.connect(db_path)
    df1.to_sql("Breast_Cancer_Original_mod", conn, index=False, if_exists="replace")
    df2.to_sql("Breast_Cancer_Wisconsin_mod", conn, index=False, if_exists="replace")
    df3.to_sql("Breast_Cancer_Metabric_mod", conn, index=False, if_exists="replace")
    print(f"Database successfully created at: {db_path}")
finally:
    conn.close()
```
