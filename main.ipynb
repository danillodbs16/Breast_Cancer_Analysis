# **Exploratory Analysis of Breast Cancer Datasets**

---

## **1. Introduction**

Breast cancer is one of the leading causes of mortality among women worldwide. This project aims to identify predictive patterns for tumor classification and survival prognosis using statistical analysis.

Two main hypotheses guide this study:

1. Tumor diagnosis (benign vs malignant) can be determined from statistical features.
2. Patient survival can be predicted using the available clinical and statistical data.

To address these hypotheses, three datasets are analyzed: two related to tumor characteristics and one focused on patient survival.

---

## **2. Data Description and Preprocessing**

Three datasets were used:

### **Breast Cancer Wisconsin (Categorical)**

- Contains clinical categorical features (values from 1 to 10)
- Target variable: tumor class (benign/malignant)

**Preprocessing steps:**
- Conversion of non-numeric values
- Handling missing data
- Standardization of target labels

---

### **Breast Cancer Wisconsin (Diagnostic - Numerical)**

- Features extracted from medical imaging (e.g., radius, texture, area)
- Target variable: diagnosis (M/B)

**Preprocessing steps:**
- Removal of irrelevant columns
- Label standardization
- Verification of data consistency

---

### **METABRIC (Survival Dataset)**

- Clinical and genomic data for survival analysis
- Includes variables such as age, tumor size, and mutation count

**Preprocessing steps:**
- Handling missing values
- Validation of clinical variables

---

### **Data Integration**

All datasets were cleaned and standardized, ensuring consistent formats and labels. The processed data was organized into a unified database to support further analysis.

---

## **3. Exploratory Data Analysis (EDA)**

Exploratory analysis was performed using statistical visualizations, including histograms, pair plots, and dimensionality reduction techniques (PCA and t-SNE).

**Key observations:**

- Both categorical and numerical datasets show strong separability between benign and malignant tumors.
- Statistical features such as radius, perimeter, and concavity are highly informative for tumor classification.
- No clear separability was observed for patient survival in the METABRIC dataset.
- Some patterns suggest a potential relationship between tumor stage and statistical features.

---

## **4. Results and Conclusions**

From the analysis:

1. Tumor classification is strongly supported by both categorical and numerical features, confirming the first hypothesis.
2. No significant evidence was found to support survival prediction using the available data, rejecting the second hypothesis.
3. There are indications that tumor stage may be predictable from certain clinical features, suggesting a possible direction for further study.
